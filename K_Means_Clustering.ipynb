K-Means_Clustering


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset= pd.read_csv("Mall_Customers.csv")
X=dataset.iloc[:,[3,4]].values
```

Using ELBOW Method To Find The Optimal Number Of Clusters


```python
from sklearn.cluster import KMeans
wcss= [ ]
for i in range(1, 11):
  Kmeans =KMeans(n_clusters= i , init = 'k-means++' , random_state=42, n_init='auto')
  Kmeans.fit(X)
  wcss.append(Kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Cluster")
plt.ylabel("WCSS")
plt.show()
```

Optimal Number is 5 as after 5 clusters the decrease in wcss becomes less steep


Now Train the Model On this 5 clusters


```python
Kmeans =KMeans(n_clusters= 5 , init = 'k-means++' , random_state=42)
y_kmeans = Kmeans.fit_predict(X)
```

```python
print(y_kmeans)
```

Generating Random Colors for the graph

```python
colors = (np.random.random(), np.random.random(), np.random.random())
```

Visulaising The Clusters

```python
for i in range(0, 5):
  colors = (np.random.random(), np.random.random(), np.random.random())
  label_name='Cluster' + str(i+1)
  plt.scatter(X[y_kmeans == i,0],X[y_kmeans == i ,1], s =100, c= colors , label=label_name)
plt.scatter(Kmeans.cluster_centers_[:,0], Kmeans.cluster_centers_[:,1], s=300 , c='black', label='Centroids')
plt.legend()
plt.title('Cluster Of Customers')
plt.xlabel('Income in k$')
plt.ylabel('Spending Score')
plt.show()
```
