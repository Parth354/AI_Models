In this case study , we are given a dataset using which we have to predict the salary of a region manager who has 2 years of experience by determining his level using Polynomial Linear Regression Model

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,-1].values
```

```python
print(X)
print(Y)
```

Training first on Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,Y)
```

Now training on Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly= poly_reg.fit_transform(X)
lin_reg_2=LinearRegression()
lin_reg_2.fit(X_poly,Y)
```

Visulalising the Linear Regression Model

```python
plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg.predict(X),color='blue')
plt.title("Truth / Bluff (Linear Regression Model)")
plt.xlabel("Level of the candidate")
plt.ylabel("Salary")
plt.show()
```

Visulaising the Polynomial Regression Model

```python
plt.scatter(X,Y,color='red')
plt.plot(X,lin_reg_2.predict(X_poly),color='blue')
plt.title("Truth / Bluff (Polynomial Regression Model)")
plt.xlabel("Level of the Candidate ")
plt.ylabel("Salary")
plt.show()
```

Predicting the salary for an input

```python
lin_reg.predict([[6.5]])
```

Predicting the salary for an input using Polynomial Regression Model

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
