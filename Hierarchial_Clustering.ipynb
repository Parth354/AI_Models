```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
X=dataset.iloc[:,[3,4]].values
```

Using Dendrogram to find the optimal Number of clusters
*italicized text*

```python
import scipy.cluster.hierarchy as sch
dendrogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Observation Point(Here Customers)')
plt.ylabel('Euclidean Distance')
plt.show()
```

Traning The Hierarchical Clustering Model


```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters= 5 , affinity='euclidean' ,linkage='ward')
y_hc=hc.fit_predict(X)
```

```python
print(y_hc)
```

Visualising the Clustering


```python
for i in range(0, 5):
  colors = (np.random.random(), np.random.random(), np.random.random())
  label_name='Cluster' + str(i+1)
  plt.scatter(X[y_hc == i,0],X[y_hc == i ,1], s =100, c= colors , label=label_name)
plt.legend()
plt.title('Cluster Of Customers')
plt.xlabel('Income in k$')
plt.ylabel('Spending Score')
plt.show()
```
