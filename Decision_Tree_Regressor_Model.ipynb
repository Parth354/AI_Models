```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset =pd.read_csv('Position_Salaries.csv')
X=dataset.iloc[:,1:-1].values
Y=dataset.iloc[:,-1].values
```

Training the Decision Tree Regression Model on the Dataset (No feature Scaling to be done)

```python
from sklearn.tree import DecisionTreeRegressor
regressor= DecisionTreeRegressor(random_state=1)
regressor.fit(X,Y)
```

Predicting The results

```python
regressor.predict([[6.5]])
```

Visualising the Result in high Resolution(Remember Decision Tress Regression is Used for multiple features and its gives an avg over an fixed interval)

```python
X_grid=np.arange(min(X),max(X), 0.1)
X_grid=X_grid.reshape((len(X_grid),1))
plt.scatter(X,Y,color='red')
plt.plot(X_grid,regressor.predict(X_grid),color='blue')
plt.title('Truth/Bluff (DecisionTreeRegressor Model)')
plt.xlabel('Position Of the Employee')
plt.ylabel('Salaries')
plt.show()
```

We get a stepwise result as Decision Tree Regressor takes an avg value over a scale and explains the working of the Decision Tree Regressor Model
